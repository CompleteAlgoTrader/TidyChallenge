# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import missingno as msno
import bamboolib
```


# Seller 1

```python
seller1 = pd.read_csv('../data/processed/seller1_cleaned.csv')
```

```python
print(seller1.columns)
```

## Page Type

```python
# Plot the page_type variable of seller1 with the count of each type on the horizontal axis. Sort the variable from highest to lowest.
sns.countplot(y='page_type', data=seller1, order=seller1['page_type'].value_counts().index)
# Vertical axis label that reads 'Page Type'
plt.ylabel('Page Type')
# Horizontal axis label that reads 'Count'
plt.xlabel('Count')
# Title of the plot
plt.title('Count of Page Type')
# Show the plot
plt.show()
```

## Location Country

```python
# Plot the top ten locations from the 'location_country' variable of seller1
sns.countplot(y='location_country', data=seller1, palette='mako_r', order=seller1['location_country'].value_counts().index[:10])
# Vertical axis label that reads 'Location Country'
plt.ylabel('Location Country')
# Horizontal axis label that reads 'Count'
plt.xlabel('Count')
# Title of the plot
plt.title('Top 10 Countries')
# Show the plot
plt.show()
```

## Total Sessions

```python
# Convert the 'hour' variable to datetime format
seller1['hour'] = pd.to_datetime(seller1['hour'])
```

```python
# Set 'hour' as the index of seller1, drop the old index
seller1 = seller1.set_index('hour')
```

```python
# Plot the average 'total_sessions' variable every day 
seller1['total_sessions'].resample('D').mean().plot()
# Vertical axis label that reads 'Total Sessions'
plt.ylabel('Total Sessions')
# Horizontal axis label that reads 'Date'
plt.xlabel('Date')
# Title of the plot
plt.title('Total Sessions')
# Show the plot
plt.show()
```
